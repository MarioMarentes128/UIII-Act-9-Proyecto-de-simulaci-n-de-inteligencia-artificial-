## Joyero, Ciente

```python
joyas = []   # La lista inicia vacía

while True:

    print("\n===================================")
    print("   💎 JOYERÍA INTELIGENTE 💎")
    print("===================================")
    print("1. Agregar joya")
    print("2. Mostrar inventario")
    print("3. Recomendar joya (IA básica)")
    print("4. Buscar joya")
    print("5. Eliminar última joya")
    print("6. Salir")
    print("===================================")

    opcion = input("Selecciona una opción: ")

    # ----------------------------------
    # AGREGAR JOYA
    # ----------------------------------
    if opcion == "1":

        nueva_joya = input("Escribe el nombre de la joya: ")
        joyas.append(nueva_joya)

        print("✅ Joya agregada correctamente")

    # ----------------------------------
    # MOSTRAR INVENTARIO
    # ----------------------------------
    elif opcion == "2":

        if joyas == []:
            print("⚠️ El inventario está vacío")

        else:
            print("\n📦 INVENTARIO ACTUAL")
            print(joyas)

    # ----------------------------------
    # RECOMENDACIÓN INTELIGENTE
    # ----------------------------------
    elif opcion == "3":

        print("\n🤖 IA DE RECOMENDACIÓN")
        print("¿Qué estilo prefieres?")
        print("1. Elegante")
        print("2. Moderno")
        print("3. Económico")

        estilo = input("Selecciona una opción: ")

        if estilo == "1":
            print("💍 La IA recomienda: Collar de diamantes")

        elif estilo == "2":
            print("⌚ La IA recomienda: Reloj inteligente dorado")

        elif estilo == "3":
            print("📿 La IA recomienda: Pulsera de plata")

        else:
            print("❌ Opción no válida")

    # ----------------------------------
    # BUSCAR JOYA
    # ----------------------------------
    elif opcion == "4":

        buscar = input("Escribe la joya que deseas buscar: ")

        if buscar in joyas:
            print("✅ La joya sí existe en el inventario")

        else:
            print("❌ La joya no fue encontrada")

    # ----------------------------------
    # ELIMINAR ÚLTIMA JOYA
    # ----------------------------------
    elif opcion == "5":

        if joyas == []:
            print("⚠️ No hay joyas para eliminar")

        else:
            eliminada = joyas.pop()
            print("🗑️ Joya eliminada:", eliminada)

    # ----------------------------------
    # SALIR
    # ----------------------------------
    elif opcion == "6":

        print("👋 Gracias por usar la Joyería Inteligente")
        break

    # ----------------------------------
    # OPCIÓN INVÁLIDA
    # ----------------------------------
    else:

        print("❌ Opción incorrecta")
```
